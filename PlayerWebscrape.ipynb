```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import string
import time

player_url = "https://basketball.realgm.com/ncaa/players/2026/"
options = webdriver.FirefoxOptions()

ublock_path = r"C:\Users\pwdev\AppData\Local\Packages\Mozilla.Firefox_n80bbvh6b1yt2\LocalCache\Roaming\Mozilla\Firefox\Profiles\ugt0ce7h.default-release\extensions\uBlock0@raymondhill.net.xpi"

driver = webdriver.Firefox()
driver.install_addon(ublock_path)

all_players = []

try:
    for letter in string.ascii_uppercase:
        url = player_url + letter
        print(f"Visiting {url}")
        driver.get(url)

        # Wait for the REAL table that exists on RealGM
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".table-striped"))
        )

        # Get rows from the REAL table
        players = pd.DataFrame(driver.find_element((By.CSS_SELECTOR, "table-striped"))

        for row in rows:
            cells = row.find_elements(By.TAG_NAME, "td")
            if len(cells) < 8:
                continue

            player = {
                "Name": cells[0].text.strip(),
                "Position": cells[1].text.strip(),
                "Height": cells[2].text.strip(),
                "Weight": cells[3].text.strip(),
                "School": cells[4].text.strip(),
                "Year": cells[5].text.strip(),
                "Home": cells[6].text.strip(),
                "HS": cells[7].text.strip()
            }
            ############# NEED TO FIX UNION OF TABLES FROM ONE LETTER TO THE NEXT #############
            all_players.append(player)

        time.sleep(0.2)

finally:
    driver.quit()

for p in all_players:
    print(p)
```
