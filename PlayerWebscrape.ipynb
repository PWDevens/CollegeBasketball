```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import string


BASE_URL = "https://basketball.realgm.com/ncaa/players/2026/"  # 2026 from your HTML
options = webdriver.FirefoxOptions()
ublock_path = "C:\\Users\\pwdev\\AppData\\Local\\Packages\\Mozilla.Firefox_n80bbvh6b1yt2\\LocalCache\\Roaming\\Mozilla\\Firefox\\Profiles\\ugt0ce7h.default-release\\extensions\\uBlock0@raymondhill.net.xpi"
driver = webdriver.Firefox()
driver.install_addon(ublock_path)

all_players = []

try:
    for letter in string.ascii_uppercase:  # A â†’ Z
        url = BASE_URL + letter
        print("Visiting:", url)

        driver.get(url)

        # Wait for the table to load
        WebDriverWait(driver, 600).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "#players tbody tr"))
        )

        # Extract rows
        rows = driver.find_elements(By.CSS_SELECTOR, "#players tbody tr")

        for row in rows:
            cells = row.find_elements(By.TAG_NAME, "td")
            if len(cells) < 8:
                continue

            player = {
                "Name": cells[0].text,
                "Position": cells[1].text,
                "Height": cells[2].text,
                "Weight": cells[3].text,
                "School": cells[4].text,
                "Year": cells[5].text,
                "Home": cells[6].text,
                "HS": cells[7].text
            }
            all_players.append(player)

except Exception as e:
    print("ERROR:", e)

finally:
    driver.quit()

# Print results
for p in all_players:
    print(p)
```
